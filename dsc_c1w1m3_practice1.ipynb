# Practice for Join Statements with SQL

Import the necessary packages and connect to the database 'data.sqlite'.

```python
# Run without change

# Libraries and database

import sqlite3
import pandas as pd

conn = sqlite3.connect('data.sqlite')

pd.read_sql(""" select * from employees """, conn)
```

1. Select the names (frist and last) of all employees in Boston

```python
# Your code here

pd.read_sql(""" SELECT firstName, lastName FROM employees JOIN offices USING(officeCode) WHERE city = 'Boston';""", conn)
```

2. Are there any offices that have zero employees?

```python
# Your code here
pd.read_sql(""" select o.officeCode, o.city from offices o LEFT JOIN  employees e using(officeCode) group by officeCode having count(e.employeeNumber) = 0;""", conn)
```

3. How many customers are there per office?

```python
# Your code here
pd.read_sql(""" select c.officeCode, count(e.salesRepEmployeeNumber), o.city from offices o join employees c using(officeCode) join customers e on c.employeeNumber = e.salesRepEmployeeNumber group by officeCode ; """, conn)
```

4. Display the names of every individual product that each employee has sold as a dataframe.

```python
# Your code here
pd.read_sql(""" 
            
           select firstName, lastName, productName from employees e join customers c on e.employeeNumber = c.salesRepEmployeeNumber
           join orders using(customerNumber)
           join orderDetails using(orderNumber)
           join products using(productCode) 
            
            
            
            ; """, conn)

```

5. Display the number of products each employee has sold
- Alphabetize the results by employee last name.
- Use the quantityOrdered column from orderDetails.
- Think about how to group the data when some employees might have the same first or last name.


```python
# Your code here

pd.read_sql(""" 
            
           select firstName, lastName, SUM(quantityOrdered) as total_products_sold from employees e join customers c on e.employeeNumber = c.salesRepEmployeeNumber
           join orders using(customerNumber)
           join orderDetails using(orderNumber)
           group by firstName, lastName
           order by lastName            
            
            
            ; """, conn)

```

6.  Display the names of employees who have sold more than 200 different products.

```python
# Your codde here
pd.read_sql(""" 
            
           select firstName, lastName, COUNT(productCode) as different_products_sold from employees e join customers c on e.employeeNumber = c.salesRepEmployeeNumber
           join orders using(customerNumber)
           join orderDetails using(orderNumber)
           group by firstName, lastName
           having different_products_sold > 200
           order by lastName      
            
            
            ; """, conn)


```
